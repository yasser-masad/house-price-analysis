```python
data['area'] = data['area'].fillna(data['area'].mean())
data['price'] = data['price'].fillna(data['price'].mean())
data['bedrooms'] = data['bedrooms'].fillna(data['bedrooms'].mean())
data['bathrooms'] = data['bathrooms'].fillna(data['bathrooms'].mean())
data['stories'] = data['stories'].fillna(data['stories'].mean())
data['parking'] = data['parking'].fillna(data['parking'].mean())

display(data.isnull().sum())
```

```python
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
def clean_outliers(data, column_list, threshold=3):
    initial_rows = len(data)
    data_filtered = data.copy()

    for column_name in column_list:
        mean_val = data_filtered[column_name].mean()
        std_val = data_filtered[column_name].std()

        lower_limit = mean_val - (threshold * std_val)
        upper_limit = mean_val + (threshold * std_val)

        data_filtered = data_filtered[
            (data_filtered[column_name] > lower_limit) &
            (data_filtered[column_name] < upper_limit)
        ]

    total_removed = initial_rows - len(data_filtered)
    print("---------------------------------------")
    print(f"✔️ Operation complete. Columns cleaned: {column_list}")
    print(f"A total of {total_removed} outlier rows were removed from the original dataset.")
    print("---------------------------------------")

    return data_filtered


columns_to_clean = ['price', 'area']

data_cleaned_final = clean_outliers(data, columns_to_clean)
```
